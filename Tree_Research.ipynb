```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("/Users/aahan_bagga/Downloads/Data for Trees Research - Sheet1.csv")
data
```

```python
#data['Inf. Rate #1(in)'] = data['start #1(inches)'] - data['end #1(inches)']
#data['Inf. Rate #2(in)'] = data['start #2(inches)'] - data['end #2(inches)']
#data['Inf. Rate #3(in)'] = data['start #3(inches)'] - data['end #3(inches)']

# Calculate the average infiltration rate for each site
#data['Average Rate'] = data[['Inf. Rate #1(in)', 'Inf. Rate #2(in)', 'Inf. Rate #3(in)']].mean(axis=1)

# Prepare data for plotting
species = data['Tree (site)']
avg_rates = data['Average Rate(in)']

# Plotting
plt.figure(figsize=(10, 6))
bars = plt.bar(species, avg_rates, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Tree (Site)', fontsize=12)
plt.ylabel('Average Infiltration Rate (inches)', fontsize=12)
plt.title('Average Infiltration Rates by Tree Species', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Annotate bars with the actual values
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height+0.02, 
             f'{height:.2f}', ha='center', fontsize=10)

# Show plot
plt.show()
```
